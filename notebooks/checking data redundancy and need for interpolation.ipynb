### 1. Check Spatial Resolution Consistency

Compare the grid resolution (latitude × longitude) of all datasets.


  --  If SST data has a finer resolution than other variables (e.g., air temperature, precipitation), you might need to           downscale other variables.
  
  --  If SST data has a coarser resolution, you may need interpolation to align it with finer datasets.


```python
lat_res = abs(sst.lat[1] - sst.lat[0])
lon_res = abs(sst.lon[1] - sst.lon[0]) 
print(f"Spatial Resolution: {lat_res}° x {lon_res}°")
```
